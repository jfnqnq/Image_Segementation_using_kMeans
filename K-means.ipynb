```python
import matplotlib as mpl
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
#imread- used to read an image file 
image = mpl.image.imread("flower.jfif")
```

```python
#imshow- used to display images
plt.imshow(image)
```

```python
"""shape- the shape typically includes the height, width,
 and the number of color channels (e.g., Red, Green, Blue for an RGB image)."""
image.shape
```

```python
"""For instance, if you have an image with shape (height, width, channels) 
and you want to flatten it into a 1D array, you could use reshape(-1)"""
X = image.reshape(-1,3)
X.shape
```

```python
#the number of clusters (groups) you want the algorithm to identify in your data
# the n_init parameter determines the number of times the algorithm will be run with different centroid initializations
kmeans = KMeans(n_clusters=2, n_init=10)
```

```python
#the algorithm adjusts its internal parameters to group the data points into clusters.
kmeans.fit(X)
```

```python
# Assign each pixel in the image the color of its cluster's centroid
segmented_img = kmeans.cluster_centers_[kmeans.labels_]

## Reshape the segmented image to its original shape
segmented_img = segmented_img.reshape(image.shape)
```

```python
# Scale pixel values to [0, 1] for proper display using plt.imshow
#Original Range: Pixel values in many images are in the range [0, 255]
#Dividing the pixel values by 255 scales them to the range [0, 1]

plt.imshow(segmented_img/255)
```
